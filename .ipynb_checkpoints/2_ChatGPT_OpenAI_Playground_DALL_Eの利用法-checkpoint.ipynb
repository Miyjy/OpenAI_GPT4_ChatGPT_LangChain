### ChatGPTのアカウント登録し方



メールアドレス、携帯電話番号、PCやインターネットがあれば利用可能

登録の流れ：

1.OpenAIのホームページへアクセス
https://openai.com/


2.登録画面 SignUpをクリック


3.メールアドレスを登録する


4.パスワードを入力する


5.Continueをクリックしてメールアドレスを認証する


6.携帯電話番号を登録する


7.登録された携帯電話にコードが送られる


8.コードを入力


9.Nextー＞Nextー＞Doneをクリックしたら登録完了（ログイン画面が表示される）



登録後のログイン流れ：

1.ログイン画面
https://platform.openai.com/apps


2.ChatGPTを選択

3.メールアドレスやPW入力したらログイン完了

### ChatGPTの画面構成



必要に応じて各機能ボタンをクリックするだけで簡単に操作ができる


入力欄について

Send a message… の欄にテキスト（日本語/英語/など）入力した言語でリアルタイムに対話形式で回答が得られる


新規質問について

話題を変えて新しい質問をしたいときは画面左上のNew chatのボタンを押せば新規の質問や回答のやり取りができる


その他操作機能について

1.Clear Conversations:チャット履歴を全消去

2.Upgrade to Plus:GPT-3.5からGPT-4.0にアップグレードする（有料版）

3.Dark Mode:ダークモードとライトモードを切り替える

4.Get Help:ヘルプ

5.Log Out:ログアウト

#### プログラム生成

        except Exception as e:
            print("ファイルの書き込みエラー:", e)

    # 使用例

## OpenAI Playgroundの利用法

### 利用料金・トークン数について



利用料金は入出力領域のテキストのトークン数で計算される、図の赤枠の数字はトークン数を表している、

OpenAI APIのサイト内で利用できる、無料枠（３か月以内の１８ドル分）で試用できる

各利用条件によって利用料金が変わる、高度な設定（プロレベル）についてはさらに詳しい利用料金がある

用途別の料金参考例

    １トークン：LLM処理のテキストの単位

    英語の場合、１単語が１～２トークン（例：Artificial)

    日本語の場合、１文字が１～２トークン（例：あ）

表：１Kトークン当たりの料金

| Mode        | 1K当たりの料金 (USD)         　| (JPY) 　　  |
|:-----------:|:-----------------------------:|:------------:|
| Input       | 0.0015     　　　　　　       | 0.21 　　　　|
| Output      | 0.002 　　　　　　            | 0.28  　　　|
| Fine-tuning | 0.0006　　　　　　　          | 0.08         |
| Embeddings  | 0.0001　　　　　　　          | 0.01         |

    為替レート：2023年7月27日時点で1 USD = 140.33 JPYで計算された数字

    単語説明

    Input/Output:入出力

    Fine-tuning :ファインチューニング

    Embedding　 :埋め込み生成


注意：

利用料金は異なる場合があるので各自のアカウント/アカウントアイコン/Manage account/Usageの順でボタンを押して確認する必要がある

お支払方法登録が必要で、利用制限設定をすることで利用の予算管理もできる

左メニューのBilling/Usage Limitで指定できる

Soft Limit:

月利用料金設置の限度が到達したときに通知をする限度設定プラン

Hard Limit：

月利用料金設置の限度が到達したときにリクエストを拒否するする設定プラン

* Username  : 任意のユーザー名

* Email address  : GitHub登録用メールアドレス

* Password  : 文字数は15文字以上、または数字と小文字を含む8文字以上のパスワードを指定する)

* Email preferences : 製品の最新情報、お知らせなどを不定期にメール受け取りたい場合にはチェックを入れる

* Verify your account  : [検証する] ボタンをクリックし、画面の指示に従って画像を選択することで、ロボットではないことを証明する

送信メールアドレス: GitHub (noreply@github.com)
